```python
#Task1:1 loading dataset using pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task1:2 Load Iris dataset from CSV file
data = pd.read_csv('iris.csv') 
```

```python
#Task1:3 Display the first 5 rows
print("Data Preview:")
print(data.head())
```

```python
#Task1:4 check datatype and missing values
print("\nData Types:")
print(data.dtypes)

print("\nMissing Values:")
print(data.isnull().sum())
```

```python
#Task1:5 droping missing values
data.fillna(data.mean(numeric_only=True), inplace=True)
```

```python
#task2:1 basic statistics
print("\nBasic Statistics:")
print("Mean:\n", data.mean(numeric_only=True))
print("\nMedian:\n", data.median(numeric_only=True))
print("\nStandard Deviation:\n", data.std(numeric_only=True))
```

```python
# Task2:2 Group by species and compute mean 
print("\nMean Values Grouped by Species:")
grouped_means = data.groupby('Species').mean(numeric_only=True)
print(grouped_means)
```

```python
# checking column names to lectify an error i had in Task2:2
print("Column Names:", data.columns.tolist())
```

```python
#Task2:3 Identify patterns or interesting findings
#  Setosa tends to have smaller petal length and width compared to Versicolor and Virginica.
#  Virginica has the largest average petal dimensions.
#  Sepal dimensions show more gap between Versicolor and Virginica.


# Correlation matrix (excluding non-numeric columns)
correlation = data.select_dtypes(include='number').corr()

print("\nCorrelation Matrix:")
print(correlation)
```

```python
#Task3:1 Line Chart
plt.figure(figsize=(8, 4))
plt.plot(data.index, data['SepalLengthCm'], marker='o')
plt.title('Sepal Length Trend Over Index')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.grid()
plt.show()
```

```python
#Task3:1 Bar Chart
plt.figure(figsize=(6, 4))
data.groupby('Species')['PetalLengthCm'].mean().plot(kind='bar', color=['#66c2a5', '#fc8d62', '#8da0cb'])
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.show()
```

```python
# Task3:1 Histogram
plt.figure(figsize=(6, 4))
plt.hist(data['SepalLengthCm'], bins=15, color='#5ab4ac', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Task3:1 Scatter Plot: Sepal length vs. petal length
plt.figure(figsize=(6, 4))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=data, palette='Set2')
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```
